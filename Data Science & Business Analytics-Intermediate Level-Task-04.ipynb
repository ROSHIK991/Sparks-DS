```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
terror.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','gname':'Group','region_txt':'Region','provstate':'State',
                       'city':'City','specificity':'Specific','latitude':'Latitude','longitude':'Longitude','summary':'Summary','attacktype1_txt':'Attacktype',
                      'targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'Killed','nwound':'Wounded','eventid':'Eventid'},inplace=True)
```

```python
terror
```

```python
terror=terror[['Year','Month','Day','Country','State','Region','City','Latitude','Longitude','Attacktype','Targettype','Killed','Wounded','Summary','Group','Weapon','motive','target1']]
```

## Barplot of the countries affected 

```python
plt.subplots(figsize=(8,8))
sns.barplot(terror['Country'].value_counts()[:10].index,terror['Country'].value_counts()[:10])
plt.title('Countries Affected',fontsize=18)
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation=40)
plt.show()
```

```python
terror[['Attacktype','Killed']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(8,8))
plt.xticks(rotation=50)
plt.title("No. of people killed",fontsize=18)
plt.xlabel('Attacktype')
plt.ylabel('No. of people')
plt.show()
```

```python
terror[['Attacktype','Wounded']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(8,8))
plt.xticks(rotation=50)
plt.title("No. of people Wounded",fontsize=18)
plt.xlabel('Attacktype')
plt.ylabel('No. of people')
plt.show()
```

```python
pd.crosstab(terror['Year'],terror['Region']).plot(kind='area',stacked=False,figsize=(8,8))
plt.xticks(rotation=30)
plt.title("Terrorist Activitiesby region in each year",fontsize=18)
plt.xlabel('Year')
plt.ylabel('No. of Attacks')
plt.show()
```

```python
Attack_per_country=terror['Country'].value_counts()[:10]
Attack_per_country
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x=terror['Attacktype'],order=terror['Attacktype'].value_counts().index)
plt.xticks(rotation=50)
plt.xlabel('Method')
plt.ylabel('Method of Attack')
plt.show()
```
